```python
import pandas as pd
```

```python
df = pd.read_csv(path+"/medical_costs.csv")
SEP='~'

to_encode = ['Sex', 'Smoker', 'Region']
continuous = ['Age', 'BMI', 'Children', 'Medical Cost']
df = pd.get_dummies(df, columns=to_encode, prefix_sep=SEP, drop_first=True)

for col in continuous:
  df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

df
```

```python
holdout = df.sample(random_state=25, frac=0.4)
train = df.iloc[df.index.difference(holdout.index)]

holdout = holdout.reset_index()

test = holdout.sample(random_state=25, frac=0.5)
validate = holdout.iloc[holdout.index.difference(test.index)]

train = train.reset_index()
validate = validate.reset_index()
test = test.reset_index()

train.to_csv("medical_train.csv", index=False)
validate.to_csv("medical_validate.csv", index=False)
test.to_csv("medical_test.csv", index=False)
```
